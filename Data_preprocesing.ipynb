## Load librarys

```python
import os

import numpy as np 
import pandas as pd
import scipy.io as sio
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
%load_ext watermark
%watermark --iversions
```

## Load data

```python
loaddata=sio.loadmat('data/feat_mat.mat')
matlab_variable = loaddata['feat_mat']
data=matlab_variable[:,:-1]
data=pd.DataFrame(data)
```

## Splitting data to training and validation parts

```python
X=data.iloc[:,:-1]# feature
Y=data.iloc[:,-1]#class
x_train,x_val,y_train,y_val=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=25,shuffle=True) #split funciton from sklearn.model_selection
y_train=keras.utils.to_categorical(y_train-1,6)
y_val=keras.utils.to_categorical(y_val-1,6)
```

## Normalization

```python
scaler = StandardScaler()
x_train_Znorm = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
joblib.dump(scaler, 'data/scaler_Ztransform.save')# save normalization parameters
```

```python
scaler = joblib.load('data/scaler_Ztransform.save')# load of parameters of normalization
x_val_Znorm = pd.DataFrame(scaler.fit_transform(x_val), columns=x_val.columns)
```
